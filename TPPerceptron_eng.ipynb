# A Neuron
A biological neuron consists of a cell body, several dendrites, and a single axon. The connections between neurons are called synapses. The neuron receives stimuli on the dendrites, and if the stimuli are sufficient, the neuron triggers (also known as activation or excitation) and sends a stimulus down its axon, which is passed on to other neurons that have synaptic connections with the excited neuron. Synaptic signals can be excitatory or inhibitory; that is, some signals may prevent a neuron from triggering instead of causing it to trigger.

# Perceptron Algorithm

The perceptron consists of a computing unit, a number of inputs, each with an input weight, and a single output. The perceptron is illustrated in the figure below.


Capture d’écran 2022-03-01 à 22.05.33.png

The inputs are usually named $ x_0, x_1, ..., x_n $ in the case of  n  inputs ($ x_0 $ being the bias input), and the output is usually named $ y $. Each input has an associated weight ($ w_i $, where $ i = 0, ..., n $), historically called synaptic weight because it represents in some way the strength of the connection from one neuron to another, but nowadays, it is generally simply called weight or input weight. For the perceptron, the output can only take one of two values, $-1 $ or $ 1 $. The bias input is always \(1\). Each input value is multiplied by its corresponding weight before it is presented to the computing unit (the dotted rectangle with rounded corners in the figure above), which roughly corresponds to the cell body of a biological neuron. The computing unit calculates the sum of the weighted inputs by their respective weights, then applies a so-called activation function, \( y = f(z) \), where \( z \) is the sum of the weighted inputs. The activation function of a perceptron is the sign function, also known as the sign function, which is \( 1 \) if the input is greater than or equal to \( 0 \) and \(-1\) otherwise. ### Perceptron Learning

The perceptron learning algorithm is a supervised learning algorithm. The model is presented with both input data and output data (also called ground truth). Think of it as a teacher presenting the question and the answer to the model, in the hope that the model will learn that a given input is associated with a corresponding output.


Let’s see how this algorithm works:

1. Initialize the weights randomly.
2. Randomly select an input and an output.
3. Compute the output value (y) using the inputs and the weights.
4. If (y) is different from the true output for this input/output pair, adjust the weights as follows:

> If (y < 0), then add (\theta x_i) to each (w_i).

> If (y > 0), then subtract (\theta x_i) from each (w_i).

5. Repeat steps 2, 3, and 4 until the perceptron correctly predicts all examples.






The perceptron has certain limitations in what it can predict, so for some sets of input/output pairs, the algorithm will not converge. However, if it is possible to propose a set of weights that allows the perceptron to represent the set of input/output pairs, then the algorithm is guaranteed to converge by finding those weights.

The arbitrary constant (\theta) is known as the learning rate and can be set in the interval ((0, 1]). The learning rate is an example of a hyperparameter, which is not a parameter adjusted by the learning algorithm.

Finally, in step 4, it may seem that all the weights will be adjusted by the same amount, but remember that the input (x_i) is not limited to taking only the two values (-1) and (1). It could very well be (0.4) for one input and (0.9) for another, so the actual weight adjustment will vary.
